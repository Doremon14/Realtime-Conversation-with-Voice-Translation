```python
# Step 1: Install required libraries
!pip install SpeechRecognition pydub gtts googletrans==4.0.0-rc1
!apt-get install ffmpeg

# Step 2: Imports
import speech_recognition as sr
from googletrans import Translator
from gtts import gTTS
from IPython.display import Audio, display
from pydub import AudioSegment
import os
import time
from google.colab import files

# Step 3: Recognize speech from uploaded file
def recognize_audio(language_code):
    print("/content/download.mp3")
    uploaded = files.upload()
    for filename in uploaded.keys():
        audio_path = filename

    if filename.endswith('.mp3'):
        sound = AudioSegment.from_mp3(audio_path)
        wav_filename = filename.replace('.mp3', '.wav')
        sound.export(wav_filename, format="wav")
        audio_path = wav_filename

    recognizer = sr.Recognizer()
    with sr.AudioFile(audio_path) as source:
        audio_data = recognizer.record(source)

    try:
        text = recognizer.recognize_google(audio_data, language=language_code)
        print(f"Recognized ({language_code}): {text}")
        return text
    except sr.UnknownValueError:
        print("Could not understand audio.")
        return ""
    except sr.RequestError:
        print("API error.")
        return ""

# Step 4: Translate text using Google Translate
def translate_text_google(text, src_lang, dest_lang):
    translator = Translator()
    translated = translator.translate(text, src=src_lang, dest=dest_lang)
    print(f"Translated ({src_lang} â†’ {dest_lang}): {translated.text}")
    return translated.text

# Step 5: Speak text
def speak(text, lang='en'):
    tts = gTTS(text=text, lang=lang)
    filename = f"tts_{int(time.time())}.mp3"
    tts.save(filename)
    display(Audio(filename, autoplay=True))

# Step 6: Main translator flow
def universal_voice_translator():
    print("Language Options:")
    print("1 - Spanish to English")
    print("2 - English to Spanish")
    print("3 - Custom source & target languages")

    choice = input("Choose an option (1/2/3): ")

    if choice == '1':
        source_lang_code = 'es-ES'
        target_lang_code = 'en'
    elif choice == '2':
        source_lang_code = 'en-US'
        target_lang_code = 'es'
    elif choice == '3':
        source_lang_code = input("Enter source speech lang code (e.g., 'fr-FR', 'ja-JP'): ")
        target_lang_code = input("Enter target lang code for translation (e.g., 'en', 'es', 'fr'): ")
    else:
        print("Invalid option.")
        return

    text = recognize_audio(language_code=source_lang_code)
    if text:
        # Convert regional codes (e.g., es-ES) to simple codes (es)
        src = source_lang_code.split('-')[0]
        translated = translate_text_google(text, src_lang=src, dest_lang=target_lang_code)
        speak(translated, lang=target_lang_code)

# Step 7: Run
universal_voice_translator()
```
